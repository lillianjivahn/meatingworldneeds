```python
#beef

import plotly.plotly as py
import pandas as pd
import csv as csv
df = pd.read_csv('file:///Users/lillianchen/Downloads/beef_kgc_2015.csv')
print(df)

scl = [[0,"rgb(166,54,3)"],[0.35,"rgb(230,85,13)"],[0.5,"rgb(253,141,60)"],
            [0.6,"rgb(253,174,107)"],[0.7,"rgb(253,208,162)"],[1,"rgb(254,237,222)"]]

data = [ dict(
        type = 'choropleth',
        locations = df['LOCATION'],
        z = df['Value'],
        text = df['COUNTRY'],
        colorscale = scl,
        autocolorscale = False,
        reversescale = True,
        marker = dict(
            line = dict (
                color = 'rgb(180,180,180)',
                width = 0.5
            ) ),
        colorbar = dict(
            autotick = False,
            title = 'Beef Consumption Density (in kg/capita)'),
      ) ]

layout = dict(
    title = '2017 Global Beef Consumption <br>Source:\
            <a href="https://data.oecd.org/agroutput/meat-consumption.htm#indicator-chart">\
            OECD Data </a>',
   geo = dict(
        showframe = True,
        showcoastlines = True,
        projection = dict(
            type = 'Mercator'
        )
    )
)

fig = dict( data=data, layout=layout )
py.iplot( fig, validate=False, filename='beef_consumption_kg_capita_2015' )
plot_url = py.iplot(fig, validate=False, filename='beef_consumption_kg_capita_2015', fileopt='new')

```
